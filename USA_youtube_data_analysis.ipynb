```python
# Correlation matrix using code found on https://stanford.edu/~mwaskom/software/seaborn/examples/many_pairwise_correlations.html
USA_youtube_df = pd.read_csv("USvideos.csv")
sns.set(style="white")

# Select columns containing continuous data
continuous_columns = USA_youtube_df[['category_id','views','likes','dislikes','comment_count']].columns

# Calculate correlation of all pairs of continuous features
corr = USA_youtube_df[continuous_columns].corr()

# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom colormap - blue and red
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, annot=True, mask=mask, cmap=cmap, vmax=1, vmin=-1,
            square=True, xticklabels=True, yticklabels=True,
            linewidths=.5, cbar_kws={"shrink": .5}, ax=ax)
plt.yticks(rotation = 0)
plt.xticks(rotation = 45)
```

Correlation between dislikes + likes vs. comment_count is similar at 0.8 and 0.73 respectively.
This is opposed to GB where dislikes had a higher correlation to comment_count, indicating GB viewers comment more on content they dislike whereas Americans comment equally on liked and disliked content.

### Fitting a model with random forest classifier

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_curve
```

```python
usa_df_train, usa_df_test = train_test_split(USA_csv, test_size=0.30, shuffle=True) #enter random seed later
```

```python
usa_cont_features = USA_csv[['likes','dislikes','views','comment_count']]
```

```python
USA_csv.dtypes
```

```python
#Need to have continuous features only
#Set up our RandomForestClassifier instance and fit to data
clf = RandomForestClassifier(n_estimators=30)
clf.fit(usa_df_train[usa_cont_features], usa_df_train['comments_disabled'])

# Make predictions
predictions = clf.predict(usa_df_test[usa_cont_features])
probs = clf.predict_proba(usa_df_test[usa_cont_features])
display(predictions)
```

```python
score = clf.score(usa_df_test[usa_cont_features], usa_df_test["comments_disabled"])
print("Accuracy: ", score)
```

```python
#Construct a confusion matrix and a ROC curve to further explore the quality of our results
get_ipython().magic('matplotlib inline')
confusion_matrix = pd.DataFrame(
    confusion_matrix(usa_df_test["comments_disabled"], predictions), 
    columns=["Predicted False", "Predicted True"], 
    index=["Actual False", "Actual True"]
)
display(confusion_matrix)

# Calculate the fpr and tpr for all thresholds of the classification
fpr, tpr, threshold = roc_curve(usa_df_test["comments_disabled"], probs[:,1])
plt.title('Receiver Operating Characteristic')
plt.plot(fpr, tpr, 'b')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```
